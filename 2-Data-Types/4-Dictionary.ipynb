#### Dictonary Comprehensions

```python
{i : i**2 for i in range(1, 11)}
```

```python
import math
d14 = {i : math.log10(i) for i in range(1, 11)}
d14
```

```python
for i in d14.keys():
    if i % 2 == 0:
        print(d14[i])
```

### Merge

```python
# Using the merge(|) operator
dict1 = {'Vivek': 24, 'Ashok':34, 'Sunil':45}

dict2 = {'Ravi':21}

dict3 = dict1 | dict2
print(dict3)
```

```python
# Using the merge(**) operator
dict1 = {'Vivek': 24, 'Ashok':34, 'Sunil':45}

dict2 = {'Ravi':21}

dict3 = {**dict1, **dict2}
print(dict3)
```
