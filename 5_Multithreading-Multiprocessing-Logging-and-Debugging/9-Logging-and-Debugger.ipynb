```python
l = [1,2,3,4,[4,5,6], 'vivek', 'jha']
```

```python
l1_int =[]
l2_str = []

for i in l:
    logging.info("This is the start of my first loop {}".format(l))
    if type(i) == list:
        for j in i:
            if type(j) == int:
                l1_int.append(j)
    elif type(i) == int:
        l1_int.append(i)
    else:
        if type(i) == str:
            l2_str.append(i)
```
